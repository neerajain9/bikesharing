```python
import pandas as pd
```

```python
# File to Load Citibike Trip Data...
tripdata = "201908-citibike-tripdata.csv"
```

```python
# 1. Create a DataFrame for the 201908-citibike-tripdata data. 
df = pd.read_csv(tripdata)
df.head()
```

```python
# 2. Check the datatypes of your columns. 
df.info()
```

```python
# 3. Convert the 'tripduration' column to datetime datatype.
df["tripduration_new"] = pd.to_datetime(df["tripduration"], unit='s')
df =df[['tripduration', 'tripduration_new', 'starttime', 'stoptime', 'start station id', 'start station name', 'start station latitude', 'start station longitude', 'end station id', 'end station name', 'end station latitude', 'end station longitude', 'bikeid', 'usertype', 'birth year', 'gender']]
df.head(5)
```

```python
# 4. Check the datatypes of your columns. 
df.info()
```

```python
# 5. Export the Dataframe as a new CSV file without the index.
df.to_csv('201908-citibike-tripdata_converted.csv', index=False)
```
